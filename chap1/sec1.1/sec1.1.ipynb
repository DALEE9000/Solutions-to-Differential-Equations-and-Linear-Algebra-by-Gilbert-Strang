```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
```

1. Draw the graph of $y = e^t$ by hand, for $âˆ’1 \le t \le 1$. What is its slope $dy/dt$ at $t = 0$? Add the straight line graph of $y = et$. Where do those two graphs cross?

```python
t = np.arange(-1,1,0.01)
y = np.exp(t)

plt.figure(figsize=(6,4))
plt.style.use(['science','scatter','dark_background'])
plt.plot(t,y, linewidth=3, color='#6A5ACD')
plt.ylim(0,)
plt.minorticks_off()
plt.savefig('chap1sec1.1ex1.eps', format='eps')
plt.show()
```
